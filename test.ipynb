# Sesiones por dia

```python
url = 'https://api.uxcam.com/v2/session/analytics'
```

```python
params = {
   'appid': appid,
   'page': 1,
   'page_size': 500,
   'filters': json.dumps([
       {
           'attribute': 'date_range',
           'operator': 'between_dates',
           'value': {
               'lower': lower,
               'upper': upper
           }
       }
    ]),
    'group_by': json.dumps([
        {'attribute': 'session_uploadedon_day'}
    ])
}
```

```python
response = requests.get(url=url, headers=headers, params=params)
json_response = response.json()
```

```python
conteo_unico_usuario = []
conteo_sesiones = []
for diction in json_response['data']:
    conteo_unico_usuario.append(diction['session_unique_user_count'])
    conteo_sesiones.append(diction['session_count'])
print(conteo_unico_usuario)
print(conteo_sesiones)
```

```python
# Sacando promedio de duración de sesión
tiempo_promedio_diario = []
for diction in json_response['data']:
    tiempo_promedio_diario.append(diction['avg_session_duration'])
print(sum(tiempo_promedio_diario) / len(tiempo_promedio_diario))
```

```python
pprint.pprint(json_response['data'])
```
